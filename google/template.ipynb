```python
import requests

# Load the Drive helper and mount
from google.colab import drive

# This will prompt for authorization.
drive.mount('/content/drive')
```

```python
# https://drive.google.com/file/d/15vL8o0X0tm7Tkt3e-8MgBtQs0pbTYCr4/view?usp=sharing

in_csv = '/content/drive/MyDrive/cs513/input1.csv'

fin = open(in_csv,'r',encoding='utf-8')

Lines = fin.readlines()
print(len(Lines))
```

```python
# read input; write output line-by-line
out_csv = '/content/drive/MyDrive/cs513/output1.csv'
with open(out_csv, 'w') as fout:
    for line in Lines: #for line in stdin:
        #print(line)
        query = "default"
        # note that this fails if there is a comma in the field
        x = line.rstrip('\n').split(',')

        # search term
        query = x[1]  

        url = "https://www.google.com/search?q=" + query + "&sourceid=chrome&ie=UTF-8"

        response = requests.get(url)
        txt = response.text.lower()

        result = ''
        line = line[:-1]  # rstrip('\n')
        if 'food' in txt:
            #print(query, "<-- found one!")
            result = line + ",1\n" 
        elif 'drink' in txt:
            #print(query, "<-- found one!")
            result = line + ",1\n"  
        #elif 'recipe' in txt:  # this captures smoked pork chop
        #    print(query, "<-- found one!")
        #    result = line + ",1\n"
        else:
            # TODO add additional search terms 
            #print(query, "<-- DO NOT EAT")
            result = line + ",0\n"     


        # write modified csv
        fout.write(result)
        print(result)
fout.close()
```
